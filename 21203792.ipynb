## Misión

Eres parte del equipo de *MercaFast*, una tienda online que procesa compras con tarjeta. El equipo tiene un módulo de pricing y checkout que ya están implementados y actualmente funcionan correctamente, pero antes de integrarlo al resto del sistema necesitan una suite de tests unitarios robusta.

### Información sobre el código entregado
El sistema tiene dos piezas principales para las cuales debes realizar pruebas unitarias: `PricingService` y `CheckoutService`.

**Estructura:** El repositorio tiene la carpeta `src` donde esta el código fuente del sistema, el archivo `pricing.py` contiene el código de la clase `PricingService` y el archivo `src/checkout.py` contiene el código de la clase `CheckoutService` y dependencias.

Además, la carpeta `tests` tiene los archivos respectivos para que implementes las pruebas unitarias de las clases `PricingService` y `CheckoutService`.

#### PricingService
Calcula el total a pagar de un carrito, incluye el subtotal, aplicación de cupón, impuestos según país, y costo de envío según país y monto.

#### CheckoutService
Orquesta el proceso de checkout:
* valida usuario y carrito
* calcula monto usando `PricingService`
* consulta un score de fraude
* intenta cobrar mediante el gateway de pagos
* si el pago resulta exitoso, crea una orden, la guarda y envia un recibo.
